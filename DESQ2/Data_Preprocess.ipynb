```python
# Import Pandas for data manipulation
import pandas as pd
```

```python
df = pd.read_csv('feature_counts.csv',sep='\t', skiprows=1)
```

```python
df.head()
```

```python
# Rename the sample columns
df.columns = [col.replace("_aligned.bam", "") for col in df.columns]

print("Updated column names:\n", df.columns)
```

```python
# Extract gene counts by selecting columns with SRR IDs only
gene_counts = df.loc[:, df.columns.str.startswith("SRR")]

gene_counts.index = df['Geneid']

gene_counts.head()
```

```python
# Save the gene counts DataFrame to a new CSV file
gene_counts.to_csv("gene_counts_only.csv")
```
